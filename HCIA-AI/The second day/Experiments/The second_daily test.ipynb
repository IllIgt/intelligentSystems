### 1、please import pandas 

```python
import pandas as pd
```

### 2、please read the Breast_cancer_data_for_ML.csv file and saved as DataFrame and name it as 'data'.

```python
df = pd.read_csv(r'Breast_cancer_data_for_ML.csv')
df
```

### 3、please drop the 'Unnamed: 32' column

```python
df.drop(['Unnamed: 32'], axis='columns', inplace = True)
print(df)
```

### 4、please to use 1,0 instead of 'M' and 'B'

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B':0})
df
```

### 5、please split the data into training data and test data

```python
from sklearn.model_selection import train_test_split
import numpy as np
y = df['diagnosis']
X = df.drop(['diagnosis'], axis=1)
X_log = np.log(X+0.001)
X_log.hist(figsize=(20, 15), color = 'c')
X_log_train, X_log_holdout, y_train, y_holdout = train_test_split(X_log, y, test_size=0.3, random_state=17)
```

### 6、please do data preprocesing 

```python
from sklearn.svm import SVC
print(X_log_train)
print(y_train)
SVM_model = SVC(
    C=1.0,
    kernel='rbf',
    degree=3,
    gamma='auto',
    coef0=0.0,
    shrinking=True,
    probability=False,
    tol=0.001,
    cache_size=200,
    class_weight=None,
    verbose=False,
    max_iter=-1,
    decision_function_shape='ovr',
    random_state=None,
)
SVM_model.fit(X_log_train,y_train)
```

### 7、please use SVM to do classification

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score
svm_pred = SVM_model.predict(X_log_holdout)
print ("Accuracy Score : ",accuracy_score(y_holdout, svm_pred) )
print ("Recall Score (how much of malignant tumours were predicted correctly) : ",recall_score(y_holdout, svm_pred))
print ("Precision Score (how much of tumours, which were predicted as 'malignant', were actually 'malignant'): ",precision_score(y_holdout, svm_pred))
```
